```python
# loads of the requirements of data 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.plot(x,y)
plt.title('Simple Plot')
plt.show()
```

```python
import pandas as pd

labels = ['A','B','C']
values = [1,2,3]

bars = plt.bar(labels, values, color='pink')

patterns = ['/','o','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.xlabel('X Axis')
plt.ylabel('y Axis')
plt.title('Bar graph')

plt.figure(figsize=(6,4))

plt.show()
```

```python
file = pd.read_csv(r"C:\Users\S Das\Downloads\covid_sample_data_updated.csv")
pivot = file.pivot_table(index='Confirmed', columns='Recovered', aggfunc='size', fill_value=0)
sns.heatmap(pivot, cmap='Reds', linecolor='white', linewidth=1)
plt.title('Heatmap Data')
plt.show()
```

```python
import matplotlib.pyplot as plt

file = pd.read_csv(r"C:\Users\S Das\Downloads\covid_sample_data_updated.csv")
sns.scatterplot(y='Confirmed', x='Country', hue='Recovered', data=file)
plt.title('Data of the Interest')
plt.show()
```

```python
sns.set_context('paper', font_scale=1.4)
sns.lmplot(x='Confirmed', y='Recovered', hue='Deaths',data=file)
plt.title('Wire Frame')
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# First plot (on ax[0])
ax[0].plot([1, 2, 3], [4, 5, 6])
ax[0].set_title('First Plot')

# Second plot (on ax[1])
ax[1].plot([1, 2, 3], [6, 5, 4])
ax[1].set_title('Second Plot')

plt.show()
```
